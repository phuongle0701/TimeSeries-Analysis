### 1. Code that draws 10,000 samples from a normal distribution with mean $<X> = 2$ and variance $<X^{2}> - <X>^{2} = 4$ 

```python
import numpy as np; 
import math; 
import matplotlib.pyplot as plt; 
plt.style.use('ggplot')
```

```python
ave  = 2; ## set average
std = np.sqrt(4); ## set standard deviation; 

N = 10000 ## number of samples; 

np.random.seed(1)
X = np.random.normal(ave, std, N)
plt.xlabel(r'$i$', fontsize = 15)
plt.ylabel(r'$X_i$', fontsize = 20)
plt.plot(X,',')
plt.show()
```

```python
plt.hist(X, bins = 20, normed = True)
x = np.arange(-10,10,0.01)
y = np.exp(-(x-ave)**2/(2*std**2))/np.sqrt(2*np.pi*std**2);
plt.xlim(-10,10)
plt.plot(x,y, color = 'b')
plt.title('Normal Distribution with mean 2 and variance 4')
plt.show()
```

## Write the code to check the equivalence of Gaussian distribution and bionomial distribution for the case of $p = 0.5$ and $M = 1000$

```python
import matplotlib.pyplot as plt; 
plt.style.use('ggplot');
### probability of the coin
p = 1/2; 
M = 10000# number of tossing

## number of experiments 
N = 1*10**4; 

ave  = M*p; 
std = np.sqrt( M*p*(1-p)); 

#print('p =',p, 'M =', M);

np.random.seed(0);
X = np.random.binomial(M,p,N); 

nmin = np.int(ave-std*5);
nmax = np.int(ave+std*5); 
nbin = nmax-nmin + 1;


### Plot the values: 
plt.hist(X,range=[nmin,nmax],bins  = nbin, normed = True )

## Gaussian distribution: 
x = np.arange(nmin,nmax,0.01/std)
y = np.exp(-(x-ave)**2/(2*std**2))/np.sqrt(2*np.pi*std**2);
plt.plot(x,y, color = 'b')
plt.xlabel(r'$n$', fontsize = 15); 
plt.ylabel(r'$P(n)$', fontsize = 15);
plt.legend([r'Gaussian', r'histogram'])
plt.show()
```

### Write a program to check the equivalence of the sum of $M$ uniform random numbers with the normal distributions. Choose the case of $M=4$

```python
M = 4; ## for the specific case

N = 100000; ## number of experiments


ave = M/2; 
std = np.sqrt(M/12); 

np.random.seed(0);
X = np.zeros(N); 
for i in range(N):
    X[i] += np.sum(np.random.rand(M)); 
```

```python
nmin = np.int(ave - 5*std); 
nmax = np.int(ave + 5*std); 
plt.hist(X, range = [nmin, nmax], bins = 50, normed = True)
x = np.arange(nmin,nmax,0.01/std)
y = np.exp(-(x-ave)**2/(2*std**2))/np.sqrt(2*np.pi*std**2)

plt.plot(x,y,color='b')
plt.ylabel(r'$P(n)$', fontsize = 15);
plt.legend([r'Gaussian', r'histogram'])
plt.show()
```

## Random Walk 

### Write a program to have probability distribution function  $𝑃(𝑙,𝑚)$  as a function of  $𝑙$  for  $𝑚=100$

```python
p = 0.5 ## the probability; 
M = 100 ## total number of random walk steps; 
N = 10000 ## number of experiments;

ave = M*(2*p-1) # average of the location L after M steps
std = np.sqrt(4*M*p*(1-p)) # standard deviation of L after M steps
print('p =',p,'M =',M)
```

```python
L = np.zeros(N); 
for i in range(N): 
    step = np.random.choice([-1,1], M)
    L[i] = np.sum(step)# calculate l after making M random steps and store to L
```

```python
nmin=np.int(ave-std*5)
nmax=np.int(ave+std*5)
nbin=np.int((nmax-nmin)/4)

plt.hist(L, range = [nmin,nmax], bins = nbin, normed = True)
x = np.arange(nmin,nmax,0.01/std); 
y = np.exp(-(x-ave)**2/(2*std**2))/np.sqrt(2*np.pi*std**2)

plt.plot(x,y,color='b')
plt.xlabel(r'$l$', fontsize = 16);
plt.ylabel(r'$P(l,m)$', fontsize = 15);
plt.legend([r'Gaussian', r'histogram'])
plt.xlim(ave-250,ave+250) # set x-range
plt.show()
```

```python
def rng(seed, a, b, M, ntotal):
    data = np.zeros(ntotal); 
    data[0] = seed
    for i in range(1, np.int(ntotal)):
         data[i] = np.mod((a*data[i-1]+b), M)
    return data/np.float(M)
```
